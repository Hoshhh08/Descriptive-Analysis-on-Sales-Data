```python
# Creating a function to automate the calculation for all columns
def calculation(data, columns):
    statistics = {}  # Created an empty dictionary to store the calculated statistics for each column

    for col in columns:  # Using a for loop to iterate through each column specified in the 'columns' list
        mean = data[col].mean()  # Calculating the mean (average) of the current column
        median = data[col].median()  # Calculating the median (middle value) of the current column
        mode = data[col].mode().iloc[0]  # Getting the mode (most frequent value) of the current column
        std = data[col].std()  # Calculating the standard deviation (spread or variation) of the current column

        statistics[col] = {  # Storing the calculated statistics for the current column in the 'statistics' dictionary
            'Mean': mean,
            'Median': median,
            'Mode': mode,
            'Std': std}

    stats_df = pd.DataFrame(statistics).T  # Converting the 'statistics' dictionary into a DataFrame for easy viewing and storing it in 'stats_df'

    return stats_df  # Returning the resulting DataFrame containing the calculated statistics for all columns
```

```python
# Calling the function to view the output
calculation(df,num_columns) # dataset stored in df and numerical values stored in num_columns

# Below are the measures of the central tendencies,
# 1. Volume - The average volume of units sold is 5.06 whereas the median value is 4(Center point of the data), with std at 4.23 showing a small
# amount of variation in the data
# 2. Avg Price - 10453 is the Average Price for per unit sold, however; std shows significantly larger variation as median value holds at 1450 only
# 3. Total Sales Value - Mean value stands at 33812 which shows the average per transaction with a median value of 5700 which means there can
# be skewness in the data. Standard Deviation of 50535 indicates a significant variation
# or a spread from the median(center part of the data) indicating a possibility of an outlier(Higher value than usual)
# 4. Discount Rate (%)	gives an average of 15.15 with a median value of 16.57 showing mean and median almost being equal and whereas standard deviation
# indicates there is a moderate variation around the median
# 5. Discount Amount's mean value is 3346 with a central value of 988 indicating a small amount of skewness to the data. Standard Deviation of 4509
# shows a bigger spread around the median of the data possibly indicating to outlier values
# 6. Net Sales Value the mean value of 30466 represents the average per transaction, but the median value of 4677 indicates there is a possible skewness
# in the data as majority of datapoints are of lower values. Standard deviation of 46358 being a significantly larger value shows the wide spread
# around the median hinting the presence of outliers.
```

### BOXPLOT

```python
# As Values for outliers are alot and bigger , we can set display options to maximum for better readability and viewing via pandas
pd.set_option('display.max_colwidth', None)  # Don't truncate column content
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)     # Show all rows
pd.set_option('display.width', 1000)        # Expand width for better readability
```

```python
def quartile(data,column):                             # Creating a function to ease the calculations for all Numerical columns
    Interquartilerange = {}                            # Empty Dictionary to add the results into
    
    for col in column:                                 # Using a for loop to iterate through each column in the dataset
        Q1 = data[col].quantile(0.25)                  # Finding the Quartile Ranges
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1                                  # Calculating IQR
        Upper_Outlier = data[col][data[col] > Q3 + 1.5*IQR].values      # Hard Coded Outlier Calculation and converted to values(as output in bool)
        Lower_Outlier = data[col][data[col] < Q1 - 1.5*IQR].values

        # Using the dictionary to add the outputs and converting outlier values into list
        
        Interquartilerange[col] = {'Q1':Q1,
                             'Q3':Q3,
                             'IQR':IQR,
                             'Upper_Outlier':Upper_Outlier.tolist(),
                             'Lower_Outlier':Lower_Outlier.tolist()}

    IQRange = pd.DataFrame(Interquartilerange).T        # Converted the entire result of the code into DataFrame for easy readability and understanding

    return IQRange                                      # Returned the Result
```

```python
Outliers = quartile(df,num_columns)   # Stored in variable Outliers if need to do slicing of columns later
Outliers
```

```python
# Showing only Outlier values for Upper and Lower bound

Outliers[['Upper_Outlier','Lower_Outlier']]
```

```python
def plotiqr(data,columns):
    # Loop through each column in the list of columns provided
    for col in columns:
        Q1 = data[col].quantile(0.25)    # Calculate the 1st quartile (25th percentile) of the column
        Median = data[col].median()      # Calculate the median (50th percentile) of the column
        Q3 = data[col].quantile(0.75)    # Calculate the 3rd quartile (75th percentile) of the column
        
        spacing = 0.75                   # Set spacing for placement of text annotations on the plot
        
        plt.figure(figsize=(8,6))        # Create a new figure with a specified size
        sns.boxplot(y = data[col],color = 'lightblue')    # Create a boxplot for the column data using seaborn
        
        # Add a text annotation for Q1 (25th percentile), positioned slightly below the box
        plt.text(0,Q1 - spacing,f'Q1: {Q1}', ha = 'left', va = 'bottom',fontsize = 10,color = 'black')

        # Add a text annotation for the Median (50th percentile), positioned on the median line
        plt.text(0,Median,f'Median: {Median}',ha = 'right', va = 'bottom',fontsize = 8,color = 'black')

        # Add a text annotation for Q3 (75th percentile), positioned slightly above the box
        plt.text(0,Q3 + spacing,f'Q3: {Q3}',ha = 'center', va = 'bottom',fontsize = 10,color = 'black')

        
        plt.title(f'Boxplot of {col}')   # Set the title of the plot using the column name
        plt.show()                       # Display the boxplot
```

```python
# Calling the function
plotiqr(df,num_columns)
```

# Standardization of Numerical Variables.

**Standardization is a data preprocessing technique used to scale numerical features so that they have the properties of a standard normal distribution.**
**This is particularly important for machine learning algorithms sensitive to the scale of input data,**
**It calculates Z-Score from the backend and standardized range values from -3 to +3.**
**Values outside of the range -3 to +3 are considered as outliers**

**Formula - z = x-mean/std**

```python
# There is a function to do standardization in python, however; will use formula to do scaling on numerical variables
# Below is to show how can do with a function

# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
#z_num = sc.fit_transform(num_columns)
#z_num
```

```python
# Let's take the Stored variable which only has Numerical Columns
num_columns.head()
```

```python
# Function to calculate Z-scores (standardization) for specified columns
def zscore(data,columns):
    z_scores = {}      # Create an empty dictionary to store the Z-scores for each column

    # Iterate through each column in the list of columns
    for col in columns:
         # Calculate the Z-score for the current column using the standardization formula
         z_calc = (data[col] - data[col].mean())/data[col].std()
         z_scores[col] = z_calc                # Add the Z-scores for the column to the dictionary

    z_score_num = pd.DataFrame(z_scores)       # Convert the dictionary of Z-scores into a pandas DataFrame for better structure and readability
    # Return the DataFrame containing standardized values for the specified columns
    return z_score_num
```

```python
z_col = zscore(df,num_columns)
z_col.head()
```

**Comparison of Data Distribution (Before and After)**

```python
# Trying the side by side comparison before giving it to function

fig, axes = plt.subplots(1,2, figsize=(15,6))
sns.histplot(df['Volume'],kde=True,color='lightblue',ax=axes[0])
axes[0].set_title('Distribution of Volume')
axes[0].set_xlabel('Volume')
axes[0].set_ylabel('Frequency')

sns.histplot(z_col['Volume'],kde=True,color='blue',ax=axes[1])
axes[1].set_title('Z-Score Distribution of Volume')
axes[1].set_xlabel('Z-Volume')
axes[1].set_ylabel('Frequency')

plt.show()
```

```python
def distcompare(data, zscoreddata, columns):
    # Create a figure with a grid of 6 rows and 2 columns for subplots
    # figsize=(16, 35) determines the size of the overall plot
    fig, axes = plt.subplots(6, 2, figsize=(16, 35))  
    
    # Loop through each column and its corresponding index
    for i, col in enumerate(columns):
        # Plot the original data distribution for the current column
        sns.histplot(data[col], kde=True, color='blue', ax=axes[i, 0])
        axes[i, 0].set_title(f'Distribution of {col}')  # Add a title for the original data
        axes[i, 0].set_xlabel(col)  # Label the x-axis with the column name
        axes[i, 0].set_ylabel('Frequency')  # Label the y-axis with 'Frequency'

        # Plot the Z-score (standardized) distribution for the current column
        sns.histplot(zscoreddata[col], kde=True, color='orange', ax=axes[i, 1])
        axes[i, 1].set_title(f'Z-Distribution of {col}')  # Add a title for the Z-score data
        axes[i, 1].set_xlabel(f'Z-{col}')  # Label the x-axis as 'Z-{column name}'
        axes[i, 1].set_ylabel('Frequency')  # Label the y-axis with 'Frequency'
    
    # Display the entire figure after all subplots are created
    plt.show()  
```

```python
# As per the Comparison between the Original Distribution and Z-Distribution,
# After scaling, the data values are transformed into Z-scores.
# The shape remains skewed (right tail still visible),
# but the range of the data is now expressed in terms of how many standard deviations each value is away from the mean.
# The Z-scaling highlights extreme values. For example, Z-scores greater than 3 or less than -3 are typically considered outliers.
# Transforming data into Z-scores improves its suitability for many analytical models
distcompare(df,z_col,num_columns)
```
